```python
import matplotlib.pyplot as plt
import numpy as np
```

### Symmetric hard limit transfer function

```python
def symmetric_hard_limit(arr, threshold=0):
    result = []
    for el in arr:
        if el >= threshold:
            result.append(1)
        else:
            result.append(-1)
    return result
```

```python
x = np.arange(start=-100, stop=101)
y = symmetric_hard_limit(x)

plt.xlabel('x')
plt.ylabel('symmetric_hard_limit(x)')
plt.plot(x,y)
plt.show()
```

### Binary step activation function

```python
def binary_step(arr, threshold=0):
    result = []
    for el in arr:
        if el >= threshold:
            result.append(1)
        else:
            result.append(0)
    return result
```

```python
x = np.arange(start=-100, stop=101)
y = binary_step(x)

plt.xlabel('x')
plt.ylabel('binary_step(x)')
plt.plot(x,y)
plt.show()
```

### Bipolar step activation function (same as Symmetric hard limit transfer function)

```python
x = np.arange(start=-100, stop=101)
y = symmetric_hard_limit(x, threshold=25)

plt.xlabel('x')
plt.ylabel('bipolar_step(x)')
plt.plot(x,y)
plt.show()
```

### Saturating linear transfer function

```python
def saturating_linear(arr):
    result = []
    for el in arr:
        if el < 0:
            result.append(0)
        elif el <= 1:
            result.append(el)
        else:
            result.append(1)
        
    return result
```

```python
x = np.arange(start=-3, stop=3, step=0.05)
y = saturating_linear(x)

plt.xlabel('x')
plt.ylabel('saturating_linear(x)')
plt.plot(x,y)
plt.show()
```

### Hyperbolic tangent sigmoid transfer function

```python
def hyperbolic_tangent_sigmoid(arr):
    result = []
    for el in arr:
        value = 2 / (1 + np.exp(-2*el)) - 1
        result.append(value)
        
    return result
```

```python
x = np.arange(start=-10, stop=10, step=0.005)
y = hyperbolic_tangent_sigmoid(x)

plt.xlabel('x')
plt.ylabel('hyperbolic_tangent_sigmoid(x)')
plt.plot(x,y)
plt.show()
```

### Log sigmoid transfer function

```python
def log_sigmoid(arr):
    result = []
    for el in arr:
        value = 1 / (1 + np.exp(-el))
        result.append(value)
        
    return result
```

```python
x = np.arange(start=-10, stop=10, step=0.005)
y = log_sigmoid(x)

plt.xlabel('x')
plt.ylabel('log_sigmoid(x)')
plt.plot(x,y)
plt.show()
```
